````python
# leap year function

def find_leap(year):
  if year%400==0:
    return True
  elif year%100==0:
    return False
  elif year%4==0:
    return True
  else:
    return False
````

## Fibonacci sequence

$$
F_{n} = 
\begin{cases}
0 & n=0\\
1 & n=1\\
F_{n-1} + F_{n-2} & n>1
\end{cases}
$$

````python
# Fibonacci Sequence

def fibo(n: int) -> int:
  if n<2:
    return n
  else:
    return fibo(n-1)+fibo(n-2)
````

````python
fibo(0)
````

````python
fibo(1)
````

````python
for i in range(2,15):
  print(fibo(i))
````

## Binet's Formula

$$
F_{n} = \frac{(1+\sqrt5)^{n}-(1-\sqrt5)^{n}}{2^{n}\sqrt5}
$$

````python
from math import sqrt

# Using binet's formula
def fibo_binet(n: int) -> int:
  return int(((1+sqrt(5))**n - (1-sqrt(5))**n)/(2**n*sqrt(5)))
````

````python
fibo_binet(6)
````

````python
s_time = time()
fibo(35)
e_time = time()
fibo_time = e_time - s_time
print(fibo_time) # Very veryveryveryveryveryveryvery slow
````

````python
s_time = time()
fibo_binet(35)
e_time = time()
binet_time = e_time - s_time
print(binet_time) # Very fast!
````

## Comprehension

### List comprehension

````python
old_list = [1,2,3,4,5]

doubled_list = []

for i in old_list:
  doubled_list.append(i*2)
doubled_list
````

````python
doubled_list = [i*2 for i in old_list]
doubled_list
````

````python
[i*2 for i in old_list]
````

````python
# list comprehension with if
old_list = [1,2,3,4,5]

doubled_list = []

for i in old_list:
  if i%2==0:
    doubled_list.append(i*2)
doubled_list
````

````python
doubled_list = [i*2 for i in old_list if i%2==0]
doubled_list
````

````python
# list comprehension with if-else
old_list = [1,2,3,4,5]

doubled_list = []

for i in old_list:
  if i%2==0:
    doubled_list.append(i*2)
  else:
    doubled_list.append(i)
doubled_list
````

````python
doubled_list = [i*2 if i%2==0 else i for i in old_list]
doubled_list
````

````python
# list comprehension with elif
old_list = [1,2,3,4,5]

doubled_list = []

for i in old_list:
  if i%2==0:
    doubled_list.append(i*2)
  elif i%3==0:
    doubled_list.append(i**2)
  else:
    doubled_list.append(i)
doubled_list
````

````python
# with ternary operator
doubled_list = [
    i*2 if i%2==0 else
    i**2 if i%3==0 else
    i for i in old_list]
````

````python
doubled_list
````

## Dictionary Comprehension

기존의 dictionary로 부터 새로운 dictionary를 생성하기 위한 표현

````python
old_dict = {1:1,2:2,3:3,4:4}

new_dict = {}
for k,v in old_dict.items():
  new_dict[k] = v*2
new_dict
````

````python
new_dict = {k:v*2 for k,v in old_dict.items()}
new_dict
````

````python
old_dict = {1:1,2:2,3:3,4:4}

new_dict = {}
for k,v in old_dict.items():
  if k%2==0:
    new_dict[k] = v*2
new_dict
````

````python
new_dict = {k:v*2 for k,v in old_dict.items() if k%2==0}
new_dict
````
