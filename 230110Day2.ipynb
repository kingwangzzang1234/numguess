## Practice - Leap year

> 윤년(閏年)은 역법을 실제 태양년에 맞추기 위해 여분의 하루 또는 월(月)을 끼우는 해이다. 태양년은 정수의 하루로 나누어떨어지지 않고, 달의 공전주기와 지구의 공전주기는 다르기 때문에 태양력에서는 하루(윤일), 태음태양력에서는 한 달(윤달)을 적절한 시기에 끼워서 이를 보정한다.

- 서력 기원 연수가 4로 나누어 떨어지는 해는 윤년으로 한다. (2020년, 2024년, 2028년, ..)
- 서력 기원 연수가 4, 100으로 나누어 떨어지는 해는 평년으로 한다. (2100년, 2200년, 2300년...)
- 서력 기원 연수가 4, 100, 400으로 나누어 떨어지는 해는 윤년으로 둔다. (1600년, 2000년,..)

## Additional Problem

방금 작성한 leap year 문제를 if statement 하나로 축약하세요



````python
is_leap = int(input('Enter year what you wanna know >>'))

if is_leap%400==0:
  print(f'{is_leap} is leap year.')
elif is_leap%100==0:
  print(f'{is_leap} is normal year.')
elif is_leap%4==0:
  print(f'{is_leap} is leap year.')
else:
  print(f'{is_leap} is normal year.')
````

````python
# Fizzbuzz 완성

#dest = int(input("Enter some num >>"))
dest = 30 # for debugging

for i in range(1, dest+1):
  if i%15==0:
    print('fizzbuzz')
  elif i%3==0:
    print('fizz')
  elif i%5==0:
    print('buzz')
  else:
    print(i)
````

````python
# 6<10 -> 더 많은 라인을 읽어내야함

#dest = int(input("Enter some num >>"))
dest = 30 # for debugging

for i in range(1, dest+1):
  if i%15==0:
    print('fizzbuzz') #2
  elif i%5==0:
    print('buzz') # 6
  elif i%3==0:
    print('fizz') # 10
  else:
    print(i) # 12
````

````python
#dest = int(input("Enter some num >>"))
dest = 30 # for debugging

for i in range(1, dest+1):
  # 1 if statement or no if statement
  if i%3==0 or i%5==0:
    print('fizz'*(i%3==0)+'buzz'*(i%5==0)) #2
  else:
    print(i) # 12
````
