## Practice(1-1)
아까 작성한 원과 구 관련 공식에 사용자가 반지름을 입력하여 이에 따른 실행결과가 나오도록 재작성하세요

````python
# Answer

pi = 3.1415
r = float(input("Enter radius> "))

jireum = round(2*r, 3)
dullet = round(2*pi*r, 3)
area = round(pi*r*r, 3)
surface = round(4*pi*r*r,3)
volume = round((4/3)*pi*r*r*r, 3)
print("jireum=", jireum)
print("dullet=", dullet)
print("area=", area)
print("surface=", surface)
print("volume=", volume)
````
