```python
class Hero:
  health = 100

  def __init__(self, name, weapon):
    self.name = name
    self.weapon = weapon
    self.inventory = []

  def attack(self):
    print(f'Attacking with {self.weapon}')
  def save_item(self, item):
    self.inventory.append(item)
  def save_item_multiple(self, item, num):
    for _ in range(num):
      self.save_item(item)
  def save_items(self, *items):
    for item in items:
      self.save_item(item)
      print(f'{item} now has been stored to inventory')

  def __del__(self):
    print(f'{self.name}: Why always me?')
```

```python
from random import randint

class Elevator:
  def __init__(self, max_cap):
    self.elevator_id = randint(1, 100000)
    self.space = []
    self.cur_cap = 0
    self.max_cap = max_cap
    self.is_up = True
  
  def close_(self):
    if self.cur_cap >= self.max_cap:
      return False
    return True

  def open_(self, weight):
    if self.cur_cap + weight <= self.max_cap:
      return True
    else:
      self.close_()
      return False
    
  def go_up(self, start, end, weight):
    if not self.is_up:
      print("elevator is going down now.")
      return
    if self.open_(weight):
      print("going up")
      self.space.append((end, weight))
    else:
      print("out of weight.")
  
  def go_down(self, start, end, weight):
    if self.is_up:
      print("elevator is goign up now")
      return
    if self.open_(weight):
      print("going down")
      self.space.append((end, weight))
      self.is_up = False
    else:
      print("out of weight.")
```

```python
class Language:
  mother_lang = "Spanish"

  def __init__(self):
    self.talk = f"I can speak {self.mother_lang}"

  @classmethod
  def talk_class_method(cls):
    return cls()

  @staticmethod
  def talk_static_method():
    return Language()

  def tell_lang(self):
    print(self.talk)

class English(Language):
  mother_lang = 'english'
```

```python
BiLingual = English()
```

```python
es = BiLingual.talk_static_method()
en = BiLingual.talk_class_method()
```

```python
es.tell_lang()
```

```python
en.tell_lang()
```
